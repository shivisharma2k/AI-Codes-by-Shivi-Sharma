```python
class State:
    def __init__(self, x, y, Id, parent, children, check):
        self.x = x
        self.y = y
        self.id = Id
        self.parent = parent
        self.children = children
        self.check = check
    def fillX(self):
        self.x = jug_1_max_cap
    def fillY(self):
        self.y = jug_2_max_cap
    def fillXFromY(self):
        if(self.y>=jug_1_max_cap-self.x):
            self.y = self.y - (jug_1_max_cap-self.x)
            self.x = jug_1_max_cap
        else:
            self.x = self.x + self.y
            self.y = 0
    def fillYFromX(self):
        if(self.x>=jug_2_max_cap-self.y):
            self.x = self.x - (jug_2_max_cap-self.y)
            self.y = jug_2_max_cap
        else:
            self.y = self.y + self.x
            self.x = 0
    def EmptyX(self):
        self.x = 0
    def EmptyY(self):
        self.y = 0
    def hasReachedFinal(self):
        if(self.x==jug_1_final_cap and self.y==jug_2_final_cap):
            return True
        return False
    def checkState(self):
        if(self.check == 0):
            check = 1
            return True
        else:
            return False
```

```python
jug_1_max_cap = 4
jug_2_max_cap = 3
jug_1_final_cap = 2
jug_2_final_cap = 0

tree = {}
s = 0 
parent = 0
root = State(0,0,s,None,[],0)
tree[s] = root
s += 1
i=0

while(i<len(tree) and tree[i].checkState()):
    i+=1  
    newNode = State(tree[parent].x,tree[parent].y,s,parent,[],0)
    if(newNode.y != 0 and newNode.x!=jug_1_max_cap):
        path = []
        newNode.fillXFromY()
        p = newNode.parent
        while(p!=None):
            path.append([tree[p].x,tree[p].y])
            p = tree[p].parent
        if([newNode.x,newNode.y] not in path):
            tree[parent].children.append(newNode)
            tree[s] = newNode
            s += 1
            
            
    newNode = State(tree[parent].x,tree[parent].y,s,parent,[],0)
    if(newNode.x != 0 and newNode.y!=jug_2_max_cap):
        path = []
        newNode.fillYFromX()
        p = newNode.parent
        while(p!=None):
            path.append([tree[p].x,tree[p].y])
            p = tree[p].parent
        if([newNode.x,newNode.y] not in path):
            tree[parent].children.append(newNode)
            tree[s] = newNode
            s += 1
            
    newNode = State(tree[parent].x,tree[parent].y,s,parent,[],0)
    if(newNode.x != 0):
        path = []
        newNode.EmptyX()
        p = newNode.parent
        while(p!=None):
            path.append([tree[p].x,tree[p].y])
            p = tree[p].parent
        if([newNode.x,newNode.y] not in path):
            tree[parent].children.append(newNode)
            tree[s] = newNode
            s += 1
            
    newNode = State(tree[parent].x,tree[parent].y,s,parent,[],0)
    if(newNode.y != 0):
        path = []
        newNode.EmptyY()
        p = newNode.parent
        while(p!=None):
            path.append([tree[p].x,tree[p].y])
            p = tree[p].parent
        if([newNode.x,newNode.y] not in path):
            tree[parent].children.append(newNode)
            tree[s] = newNode
            s += 1
            
    newNode = State(tree[parent].x,tree[parent].y,s,parent,[],0)
    if(newNode.x != jug_1_max_cap):
        path = []
        newNode.fillX()
        p = newNode.parent
        while(p!=None):
            path.append([tree[p].x,tree[p].y])
            p = tree[p].parent
        if([newNode.x,newNode.y] not in path):
            tree[parent].children.append(newNode)
            tree[s] = newNode
            s += 1
            
    newNode = State(tree[parent].x,tree[parent].y,s,parent,[],0)
    if(newNode.y != jug_2_max_cap):
        path = []
        newNode.fillY()
        p = newNode.parent
        while(p!=None):
            path.append([tree[p].x,tree[p].y])
            p = tree[p].parent
        if([newNode.x,newNode.y] not in path):
            tree[parent].children.append(newNode)
            tree[s] = newNode
            s += 1
   
    parent +=1
```

```python
for i in tree:
    print(str(i) + " - " + str(tree[i].parent) + " : " + str(tree[i].x) + "," + str(tree[i].y))
```

```python
allPaths=[]
check = 1
count = 1
for i in tree:
    if(tree[i].x ==jug_1_final_cap and tree[i].y == jug_2_final_cap):
        path = []
        path.append([tree[i].x,tree[i].y])
        path.append("->")
        parent = tree[i].parent
        while(parent!=None):
            path.append([tree[parent].x,tree[parent].y])
            path.append("->")
            parent = tree[parent].parent
        for i in path:
            if(path.count(i)>1 and i != "->"):
                check = 0
                break
        if(path not in allPaths and check):
            allPaths.append(path)
            print(count, end=") ")
            for j in reversed(path):
                print(j, end =" ")
            print("\n")
            count += 1
```
