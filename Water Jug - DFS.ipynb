```python
stack = []
jug_1_max_cap = 3
jug_2_max_cap = 4
jug_1_final_cap = 0
jug_2_final_cap = 2

class State:
    def __init__(self, x, y, Id, parent, children, case):
        self.x = x
        self.y = y
        self.id = Id
        self.parent = parent
        self.children = children
        self.case = case
    def fillX(self):
        self.x = jug_1_max_cap
    def fillY(self):
        self.y = jug_2_max_cap
    def fillXFromY(self):
        if(self.y>=jug_1_max_cap-self.x):
            self.y = self.y - (jug_1_max_cap-self.x)
            self.x = jug_1_max_cap
        else:
            self.x = self.x + self.y
            self.y = 0
    def fillYFromX(self):
        if(self.x>=jug_2_max_cap-self.y):
            self.x = self.x - (jug_2_max_cap-self.y)
            self.y = jug_2_max_cap
        else:
            self.y = self.y + self.x
            self.x = 0
    def EmptyX(self):
        self.x = 0
    def EmptyY(self):
        self.y = 0
    def hasReachedFinal(self):
        if(self.x==jug_1_final_cap and self.y==jug_2_final_cap):
            return True
        return False

def printTree(tree):
    for i in tree:
        print(str(i) + " : " + str(tree[i].x) + " " + str(tree[i].y) + "    " + str(tree[i].parent) + " " + str(tree[i].children) + "  - " + str(tree[i].case))
    print()
```

```python
tree = {}
root = State(0,0,0,None,[],0)
tree[0] = root
stack.append([0,0])        

s = 1
parent = 0

printTree(tree)
print()

while(parent != None):
    
    newNode = State(tree[parent].x, tree[parent].y, s, parent,[], 0)
    if(tree[parent].case < 1):
        tree[parent].case += 1
        if(newNode.y != jug_2_max_cap):
            newNode.fillY()
            if([newNode.x, newNode.y] not in stack):
                tree[parent].children.append(s)
                tree[s] = newNode
                stack.append([newNode.x, newNode.y])        
                parent = s
                s += 1
                printTree(tree)
                if(newNode.hasReachedFinal()):
                    break   
                continue
    
    newNode = State(tree[parent].x, tree[parent].y, s, parent,[], 0)
    if(tree[parent].case < 2):
        tree[parent].case += 1
        if(newNode.x != jug_1_max_cap):
            newNode.fillX()
            if([newNode.x, newNode.y] not in stack):
                tree[parent].children.append(s)
                tree[s] = newNode
                stack.append([newNode.x, newNode.y])        
                parent = s
                s += 1
                printTree(tree)
                if(newNode.hasReachedFinal()):
                    break   
                continue
    
    newNode = State(tree[parent].x,tree[parent].y,s,parent,[],0)
    if(tree[parent].case < 3):
        tree[parent].case += 1
        if(newNode.y != 0):
            newNode.EmptyY()
            if([newNode.x, newNode.y] not in stack):
                tree[parent].children.append(s)
                tree[s] = newNode
                stack.append([newNode.x, newNode.y])  
                parent = s
                s += 1
                printTree(tree)
                if(newNode.hasReachedFinal()):
                    break   
                continue
    
    newNode = State(tree[parent].x,tree[parent].y,s,parent,[],0)
    if(tree[parent].case < 4):
        tree[parent].case += 1
        if(newNode.x != 0):
            newNode.EmptyX()
            if([newNode.x, newNode.y] not in stack):
                tree[parent].children.append(s)
                tree[s] = newNode
                stack.append([newNode.x, newNode.y])  
                parent = s
                s += 1
                printTree(tree)
                if(newNode.hasReachedFinal()):
                    break   
                continue
    
    newNode = State(tree[parent].x,tree[parent].y,s,parent,[],0)
    if(tree[parent].case < 5):
        tree[parent].case += 1
        if(newNode.x != 0 and newNode.y!=jug_2_max_cap):
            newNode.fillYFromX()
            if([newNode.x, newNode.y] not in stack):
                tree[parent].children.append(s)
                tree[s] = newNode
                stack.append([newNode.x, newNode.y])  
                parent = s
                s += 1
                printTree(tree)
                if(newNode.hasReachedFinal()):
                    break   
                continue
    
    newNode = State(tree[parent].x,tree[parent].y,s,parent,[],0)
    if(tree[parent].case < 5):
        tree[parent].case += 1
        if(newNode.y != 0 and newNode.x!=jug_1_max_cap):
            newNode.fillXFromY()
            if([newNode.x, newNode.y] not in stack):
                tree[parent].children.append(s)
                tree[s] = newNode
                stack.append([newNode.x, newNode.y])  
                parent = s
                s += 1
                printTree(tree)
                if(newNode.hasReachedFinal()):
                    break   
                continue
    
    stack.pop()    
    parent -= 1
```
