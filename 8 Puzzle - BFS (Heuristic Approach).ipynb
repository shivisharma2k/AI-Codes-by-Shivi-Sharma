```python
# [1, '', 6, 5, 3, 8, 4, 2, 7]
# [7, 5, 2, 3, '', 8, 1, 6, 4]
# [6, 5, 3, 4, 8, 1, 2, '', 7]
# [4, 7, 8, 2, 3, 6, '', 5, 1]
# [5, 3, 7, 1, 2, 6, '', 8, 4]
# [1, 8, 2, 3, 5, 7, '', 6, 4]
```

```python
import random
input_matrix = [1,2,5,3,4,8,6,7,'']
random.shuffle(input_matrix)
final_matrix = ['',1,2,3,4,5,6,7,8]

tree = {}
path = [input_matrix]
class Puzzle:
    def __init__(self, matrix, id, parent, accuracy, children, child_accuracy):
        self.matrix = matrix
        self.id = id
        self.parent = parent
        self.accuracy = accuracy
        self.children = children
        self.child_accuracy = child_accuracy
        
    def shiftUp(self,s):
        index = self.matrix.index('')
        if(index-3>=0 and index-3<=8):
            self.matrix[index], self.matrix[index-3] = self.matrix[index-3], self.matrix[index]
            if(self.matrix not in path):
                self.accuracy = self.placedAccuracy()
                tree[parent].children.append(s)
                tree[parent].child_accuracy.append(self.accuracy)
                tree[s] = self
                path.append(self.matrix)
                return(s+1)
        return s
            
    def shiftLeft(self,s):
        index = self.matrix.index('')
        if(index-1>=0 and index-1<=8 and index%3>=1):
            self.matrix[index], self.matrix[index-1] = self.matrix[index-1], self.matrix[index]
            if(self.matrix not in path):
                self.accuracy = self.placedAccuracy()
                tree[parent].children.append(s)
                tree[parent].child_accuracy.append(self.accuracy)
                tree[s] = self
                path.append(self.matrix)
                return(s+1)
        return s
            
    def shiftDown(self,s):
        index = self.matrix.index('')
        if(index+3>=0 and index+3<=8):
            self.matrix[index], self.matrix[index+3] = self.matrix[index+3], self.matrix[index]
            if(self.matrix not in path):
                self.accuracy = self.placedAccuracy()
                tree[parent].children.append(s)
                tree[parent].child_accuracy.append(self.accuracy)
                tree[s] = self
                path.append(self.matrix)
                return(s+1)
        return s
        
    def shiftRight(self,s):
        index = self.matrix.index('')
        if(index+1>=0 and index+1<=8 and index%3<=1):
            self.matrix[index], self.matrix[index+1] = self.matrix[index+1], self.matrix[index]
            if(self.matrix not in path):
                self.accuracy = self.placedAccuracy()
                tree[parent].children.append(s)
                tree[parent].child_accuracy.append(self.accuracy)
                tree[s] = self
                path.append(self.matrix)
                return(s+1)
        return s
            
    def placedAccuracy(self):
        count=0
        for i in range(len(self.matrix)):
            if(self.matrix[i]==final_matrix[i]):
                count += 1
        return count
    
    def hasReachedFinal(self):
        if(self.matrix == final_matrix):
            return True
        return False
    
def printTree(tree):
    for i in tree:
        print(str(i) + " - " + str(tree[i].matrix) + " (" + str(tree[i].accuracy) + ") -:-" 
              + str(tree[i].children) + " , " + str(tree[i].child_accuracy))
```

```python
import copy
root = Puzzle([7, 4, '', 6, 5, 3, 1, 8, 2],0,None,0,[],[])
tree[0] = root
s = 1
parent  = 0

print("Input : " + str([7, 4, '', 6, 5, 3, 1, 8, 2]) + "\nFinal : " + str(final_matrix) + "\n")

while(True):
    newNode = Puzzle(copy.deepcopy(tree[parent].matrix),s,parent,0,[],[])
    s = newNode.shiftUp(s)
    if(newNode.hasReachedFinal()):
        break
        
    newNode = Puzzle(copy.deepcopy(tree[parent].matrix),s,parent,0,[],[])
    s = newNode.shiftLeft(s)
    if(newNode.hasReachedFinal()):
        break
        
    newNode = Puzzle(copy.deepcopy(tree[parent].matrix),s,parent,0,[],[])
    s = newNode.shiftDown(s)
    if(newNode.hasReachedFinal()):
        break
        
    newNode = Puzzle(copy.deepcopy(tree[parent].matrix),s,parent,0,[],[])
    s = newNode.shiftRight(s)
    if(newNode.hasReachedFinal()):
        break
        

    if(len(tree[parent].child_accuracy)!=0):
        i = tree[parent].child_accuracy.index(max(tree[parent].child_accuracy))
        tree[parent].child_accuracy.pop(i)
        parent = tree[parent].children.pop(i)
    else:
        parent = tree[parent].parent
        while(len(tree[parent].child_accuracy)==0):
            parent = tree[parent].parent
        i = tree[parent].child_accuracy.index(max(tree[parent].child_accuracy))
        tree[parent].child_accuracy.pop(i)
        parent = tree[parent].children.pop(i)
        
    if(s==50000):
        print("*****LIMIT EXCEEDED***** \n\n")
        break
printTree(tree)
```
