```python
class Boat:
    def __init__(self, m, c, Id, parent, children, side, left, right):
        self.x = m
        self.y = c
        self.Id = Id
        self.parent = parent
        self.children = children
        self.side = side
        self.left = left
        self.right = right
    def addOneMissionary(self):
        self.x += 1
        if(self.side == 'R'):
            self.left[0] -= 1
            self.right[0] += 1
        elif(self.side == 'L'):
            self.right[0] -= 1
            self.left[0] += 1
    def addOneCannibal(self):
        self.y += 1
        if(self.side == 'R'):
            self.left[1] -= 1
            self.right[1] += 1
        elif(self.side == 'L'):
            self.right[1] -= 1
            self.left[1] += 1
    def addOneEach(self):
        self.x += 1
        self.y += 1
        if(self.side == 'R'):
            self.left[0] -= 1
            self.right[0] += 1
            self.left[1] -= 1
            self.right[1] += 1
        elif(self.side == 'L'):
            self.right[0] -= 1
            self.left[0] += 1
            self.right[1] -= 1
            self.left[1] += 1
    def addTwoMissionary(self):
        self.x += 2
        if(self.side == 'R'):
            self.left[0] -= 2
            self.right[0] += 2
        elif(self.side == 'L'):
            self.right[0] -= 2
            self.left[0] += 2
    def addTwoCannibal(self):
        self.y += 2
        if(self.side == 'R'):
            self.left[1] -= 2
            self.right[1] += 2
        elif(self.side == 'L'):
            self.right[1] -= 2
            self.left[1] += 2
    def isValid(self):
        if(self.checkLeft() and self.checkRight()):
            return True
        return False
    def checkLeft(self):
        if(self.left[0]>=0 and self.left[1]>=0 and self.left[0]<=3 and self.left[1]<=3):
            if(self.left[1]>0):
                if(self.left[0]<= self.left[1]):
                    return True
            else:
                return True
        return False
    def checkRight(self):
        if(self.right[0]>=0 and self.right[1]>=0 and self.right[0]<=3 and self.right[1]<=3):
            if(self.right[1]>0):
                if(self.right[0]<= self.right[1]):
                    return True
            else:
                return True
        return False
    def hasReachedFinal(self):
        if(self.right[0]==3 and self.right[1]==3):
            return True
        return False
```

```python
root = Boat(0,0,0,None,[],'L',[3,3],[0,0])
tree = {}
tree[0] = root
identifier = 1
parent  = 0
flag = 0

while(True):
    if(tree[parent].side == 'L'):
        side = 'R'
    else:
        side = 'L'
    
    newNode = Boat(tree[parent].x,tree[parent].y,identifier,parent,[],side,[tree[parent].left[0],tree[parent].left[1]],[tree[parent].right[0],tree[parent].right[1]])
    newNode.addOneMissionary()
    if(newNode.isValid()):
        tree[parent].children.append(newNode)
        tree[identifier] = newNode
        identifier += 1
        if(newNode.hasReachedFinal()):
            break
    newNode = Boat(tree[parent].x,tree[parent].y,identifier,parent,[],side,[tree[parent].left[0],tree[parent].left[1]],[tree[parent].right[0],tree[parent].right[1]])
    newNode.addOneCannibal()
    if(newNode.isValid()):
        tree[parent].children.append(newNode)
        tree[identifier] = newNode
        identifier += 1
        if(newNode.hasReachedFinal()):
            break
    newNode = Boat(tree[parent].x,tree[parent].y,identifier,parent,[],side,[tree[parent].left[0],tree[parent].left[1]],[tree[parent].right[0],tree[parent].right[1]])
    newNode.addOneEach()
    if(newNode.isValid()):
        tree[parent].children.append(newNode)
        tree[identifier] = newNode
        identifier += 1
        if(newNode.hasReachedFinal()):
            break
    newNode = Boat(tree[parent].x,tree[parent].y,identifier,parent,[],side,[tree[parent].left[0],tree[parent].left[1]],[tree[parent].right[0],tree[parent].right[1]])
    newNode.addTwoMissionary()
    if(newNode.isValid()):
        tree[parent].children.append(newNode)
        tree[identifier] = newNode
        identifier += 1
        if(newNode.hasReachedFinal()):
            break
    newNode = Boat(tree[parent].x,tree[parent].y,identifier,parent,[],side,[tree[parent].left[0],tree[parent].left[1]],[tree[parent].right[0],tree[parent].right[1]])
    newNode.addTwoCannibal()
    if(newNode.isValid()):
        tree[parent].children.append(newNode)
        tree[identifier] = newNode
        identifier += 1
        if(newNode.hasReachedFinal()):
            break
            
    parent += 1
```
